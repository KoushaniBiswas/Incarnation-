```
import numpy as np
#importing numpy library and renaming it
```

```
import matplotlib.pyplot as plt
image = plt.imshow(data)
plt.show()
#importing matlplotlib library to create a heat map of the data
```

```
#looking at the average inflammation over time - using ave_inflammation variable
ave_inflammation = np.mean(data, axis=0) #finding out the mean data
ave_plot = plt.plot(ave_inflammation) #using matplotlib to plot the mean data
plt.show
```

```
min_plot = plt.plot(np.min(data, axis = 0))
plt.show()
#min value has uniform step function
```

```
import numpy as np
import matplotlib.pyplot as plt
#loadtxt readsour data
data = np.loadtxt(fname = 'inflammation-05.csv', delimiter =',')
fig = plt.figure(figsize=(10.0, 3.0)) #plt.figure() creates a space in which weplace all of our plots
#figsize parameter tells python how big the space would be
##each subplot placed using addsubplot
axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)
#three parameters are taken 
#first - how many total rows of subplot present
#second - total number ofsubplot column
#third - which subplot our variable is refering to
# each subplot as three axes - and stored in different variables
axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis = 0))

axes2.set_ylabel('max')
axes2.plot(np.max(data, axis = 0))

axes3.set_ylabel('min')
axes3.plot(np.min(data, axis = 0))

fig.tight_layout()

plt.savefig('inflammation5.png')
plt.show()
```

```
#creating a plot using the standard deviation of all the data
#np.std gives the standard deviation of the data sets
std_plot = plt.plot(np.std(data, axis = 0))
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
#loadtxt readsour data
data = np.loadtxt(fname = 'inflammation-05.csv', delimiter =',')
fig = plt.figure(figsize=(10.0, 3.0)) #plt.figure() creates a space in which weplace all of our plots
#figsize parameter tells python how big the space would be
##each subplot placed using addsubplot
axes1 = fig.add_subplot(3, 1, 1)
axes2 = fig.add_subplot(3, 1, 2)
axes3 = fig.add_subplot(3, 1, 3)
#three parameters are taken 
#first - how many total rows of subplot present
#second - total number ofsubplot column
#third - which subplot our variable is refering to
# each subplot as three axes - and stored in different variables
axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis = 0))

axes2.set_ylabel('max')
axes2.plot(np.max(data, axis = 0))

axes3.set_ylabel('min')
axes3.plot(np.min(data, axis = 0))

fig.tight_layout()

plt.savefig('inflammation5.png')
plt.show()
```
