STACKING ARRAYS

```
import numpy  
A = numpy.array([[1,2,2], [4,5,6], [7,8,9]])
print('A = ')
print(A)

B = numpy.hstack([A, A])
print('B = ')
print(B)

C = numpy.vstack([A, A])
print('C = ')
print(C)
```
